2. Adım: Açık kaynak bir veri kümesini indirip (kodla), dataFrame'e çevirip, yazdır.

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd

# indireceğimiz dosyanın yolu
file_path = "~/merdandev/mlcourse/ml-course-homeworks/odev1/data/fromkaggle.csv"

# kaggledan indir
download_dir = kagglehub.dataset_download(
  "zynicide/wine-reviews"
)


# veri setinin yolunu oluştur
csv_path = f"{download_dir}/winemag-data-130k-v2.csv"

# dosyanın yolunu yazdır
print("dosya yolu:", csv_path)

# veri setini yükle
df = pd.read_csv(csv_path)

# veri setinin ilk 10 kaydını yazdır
print("ilk 10 kayıt:", df.head(10))
```

3. Adım: 1. adımdaki dosyayı oku ve yazdır

4. adım: 2 ve 3teki okuduğumuz veriler görselleştirilecek. 4er öznitelik seçilecek

```python
#fromkaggle.csv dosyasındaki verilerden 4 öznitelik seçip görselleştir

import matplotlib.pyplot as plt

# 4 öznitelik seç
df_selected = df[["country", "designation", "points", "price"]] #seçtiğimiz öznitelikleri atadık
# görselleştir
df_selected.plot(kind='scatter', x='price', y='points', alpha=0.5) #scatter çizdirme
plt.title('Price vs Points') #başlık
plt.xlabel('Price') #x ekseni
plt.ylabel('Points')    #y ekseni
plt.show()  #görselleştir



#trenler.xlsx dosyasındaki verilerden 4 öznitelik seçip görselleştir
df_tren_selected = df_tren[["Model", "İşletici", "Maks. Hız (İşletim)", "Kullanıma Girdiği Yıl"]] #seçtiğimiz öznitelikleri atadık
# görselleştir
df_tren_selected.plot(kind='scatter', x='Kullanıma Girdiği Yıl', y='Maks. Hız (İşletim)', alpha=0.5) #scatter çizdirme
plt.title('Kullanıma Girdiği Yıl vs Maks. Hız (İşletim)') #başlık
plt.xlabel('Kullanıma Girdiği Yıl') #x ekseni
plt.ylabel('Maks. Hız (İşletim)')   #y ekseni
plt.show()  #görselleştir
```
